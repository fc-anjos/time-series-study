```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def parser(x):
    print(x)
    return datetime.strptime(x, '%Y-m')

sales = pd.read_csv('./data/sales-cars.csv')
sales['Month'] = pd.to_datetime(sales['Month'], format='%Y-%m')
sales.index = sales['Month']
sales = sales.drop(columns=['Month'])
```

```python
sales.head()
```

```python
sales.plot()
```

## Is it stationary?
No: mean, variance and covariance is not constant.

### Converting series to stationary
We need to find the difference

```python
sales_diff = sales.diff(periods=1)[1:] ## Drop first nan
## Integrated of order 1, denoted by d (for diff), one of the parameter of ARIMA model
```

```python
plot_acf(sales)
```

```python
plot_acf(sales_diff) #Quickly dropping, means stationary data
```

```python
sales_diff.plot()
```
