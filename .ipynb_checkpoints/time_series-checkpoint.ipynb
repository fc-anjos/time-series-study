```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import itertools
import numpy as np
import math
import warnings

from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error

from statsmodels.tsa.arima_model import ARIMA
```

```python
## Divide into train and test data
## TODO: Automate division to be 80/20
X = sales.values
train = X[:27]
test = X[26:]
predictions = []
```

# ARIMA model

```python
#p = periods taken into account for autoregressive
#d = order of integration, how many times difference is done
#q = moving average
model_arima = ARIMA(train, order=(3, 1, 1)) #order(p, d, q)
model_arima_fit = model_arima.fit()
print(model_arima_fit.aic) #The lower this is, the better fitting
```

```python
predictions_arima = model_arima_fit.forecast(steps=10)[0]
plt.plot(test)
plt.plot(predictions_arima, color='red')
```

## Automation of lowest AIC finding (best params)

```python
p=d=q=range(0,5)
pdq = list(itertools.product(p, d, q)) ## Permutation

warnings.filterwarnings('ignore')

best_aic = math.inf

for params in pdq:
    try:
        model_arima = ARIMA(train, order=params) #order(p, d, q)
        model_arima_fit = model_arima.fit()
        current_aic = model_arima_fit.aic

    except ValueError:
        continue
    except np.linalg.LinAlgError:
        continue

    if current_aic < best_aic:
        best_aic = current_aic
        best_params = params
```

```python
model_arima = ARIMA(train, order=best_params) #order(p, d, q)
model_arima_fit = model_arima.fit()
predictions_arima = model_arima_fit.forecast(steps=10)[0]
plt.plot(test)
plt.plot(predictions_arima, color='red')
```
